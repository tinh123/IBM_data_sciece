# I. Introduction to data visualization

### 1. Why build visuals? 


- Data visualization is a way to show a complex data in a form that is graphical and easy to understand. This can be especially useful when one is trying to explore the data and getting acquainted with it.
- Plots and graphs can be very effective in conveying a clear description of the data especially when disclosing findings to an audience or sharing the data with other peer data scientists.
- They can be very valuable when it comes to supporting any recommendations you make to clients managers or other decision-makers in your field.

### 2. Practices  

    Darkhorse Analytics specializes in quantitative consulting in several areas including data visualization and geo spatial analysis. Their approach when creating a visual revolves around three key points: less is
more effective, it is more attractive, and it is more impactive.
In other words, any
feature or design you incorporate in your plot to make it more attractive or
pleasing should support the message that the plot is meant to get across and not
distract from it.

# I. Introduction to matplotlib


Matplotlib's architecture is composed of three main layers: 
- The back-end layer
- The artist layer where much of the heavy lifting happens and is usually the appropriate programming paradigm when writing a web application server, or a UI application, or perhaps a script to be shared with other developers.
- The scripting layer, which is the appropriate layer for everyday purposes and is considered a lighter scripting interface to simplify common tasks and for a quick and easy generation of graphics and plots. 

http://aosabook.org/en/matplotlib.html

- One limitation of this backend is that you cannot modify a figure once it's rendered. 
So after rendering the above figure, there is no way for us to add, for example, a figure title or label its axes. You will need to generate a new plot and add a title and the axes labels before calling the show
function. 

```python
plt.plot(5,5,'o')
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Plotting example")

plt.show()
```
